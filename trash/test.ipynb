```python
data_files = {
    "train": "drugsComTrain_raw.tsv",
    "test": "drugsComTest_raw.tsv"
}

drug_ds = load_dataset("csv", data_files=data_files, delimiter="\t")
```

```python
for split in drug_ds:
    print(split, "->", len(drug_ds[split]) == len(drug_ds[split].unique("Unnamed: 0")))
```

```python
drug_ds = drug_ds.rename_column(original_column_name="Unnamed: 0", new_column_name="id")
```

```python
drug_ds
```

```python
from pprint import pprint
for split in drug_ds:
    out1 = drug_ds[split].unique("drugName")
    out2 = drug_ds[split].unique("condition")
    pprint(f"{split} -> {len(out1), len(out2)}")
    pprint("None" in out2)
```

```python
def lowercase_condition(example):
    return {"condition": example["condition"].lower()}
drug_ds = drug_ds.filter(lambda x: x["condition"] is not None)
drug_ds = drug_ds.map(lowercase_condition)
```

```python
drug_ds
```

```python
drug_ds["train"]["condition"][:5]
```

```python
def compute_review_length(example):
    return {"review_length": len(example["review"].split())}

drug_ds = drug_ds.map(compute_review_length)
```

```python
drug_ds
```

```python
drug_ds["train"][0]
```

```python
drug_ds["train"].sort("review_length")[-3:]
```

```python
drug_ds = drug_ds.filter(lambda x: x["review_length"]>30)
drug_ds
```

```python
drug_ds["train"].sort("review_length")[:3]
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

def tokenize_func(examples):
    return tokenizer(examples["review"], truncation=True)
```

```python
%%time 
tokenized_drug_ds = drug_ds.map(tokenize_func, batched=True)
```

# `Slow Tokenizer`

```python
slow_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased", use_fast=False)
```

```python
%%time
def slow_tokenize_func(example):
    return slow_tokenizer(example["review"], truncation=True)

slow_tds = drug_ds.map(slow_tokenize_func, batched=True, num_proc=8)
```
